## 3: Exploratory Data Analysis (EDA)

## Scatter Plot (Relationship Between Features)
A scatter plot uses dots to represent values for two different numeric variables.


```python
# Create a new figure with a fixed size for clarity
plt.figure(figsize=(6,4))

# Scatter plot showing relationship between sepal length and sepal width
# Points are colored based on flower species
sns.scatterplot(
    x="sepal_length",      # X-axis feature
    y="sepal_width",       # Y-axis feature
    hue="species",         # Color differentiation by species
    data=df                # Iris dataset
)

# Add labels to both axes for better understanding
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")

# Add a clear and descriptive title
plt.title("Sepal Length vs Sepal Width by Species")

# Customize and display the legend
plt.legend(title="Species")

# Show the plot
plt.show()
```

## Histogram
A histogram is a graphical representation used in statistics to show the frequency distribution of a continuous, numerical dataset

```python
# Create histograms for all numerical features
# This helps in understanding data distribution and skewness
df.hist(
    figsize=(8,6),   # Size of the overall histogram grid
    bins=20            # Number of bins for frequency distribution
)

# Add a main title for all histograms
plt.suptitle("Distribution of Numerical Features", fontsize=14)

# Display histograms
plt.show()
```

## Box Plot
 A box plot uses boxes and lines to depict the distributions of one or more groups of numeric data.
 Box limits indicate the range of the central 50% of the data, with a central line marking the median value

```python
# Create a box plot for numerical features
# Box plots help identify spread and potential outliers

plt.figure(figsize=(8,6))

# Drop categorcal column before plotting
sns.boxplot(
    df.drop("species", axis=1)

)

# Add title
plt.title("Box plot showing Feature spread and Outliers")

# Display Box plot
plt.show()
```
